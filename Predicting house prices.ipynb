```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_data_08-02-2021.csv')
df.head()
```

```python
df.replace('Договорная', 35000, inplace = True)
```

## Vazifani CRSIP-DM Metolodgiyasi yordamida bajaring.
<img src="https://i.imgur.com/dzZnnYi.png" alt="CRISP-DM" width="800"/>

```python
import sklearn

df[df['size'] == 'Площадьземли:1сот'].index
df.drop([5347], inplace = True)

from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size = 0.2, random_state = 35)

x_train = train_set.drop('price', axis = 1)
y = train_set['price'].copy()
x_num = x_train.drop(['location', 'district'], axis = 1)

from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_pipeline= Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('std_scaler', StandardScaler())
])

from sklearn.compose import ColumnTransformer
num_attribs = list(x_num)
cat_attribs = ['location', 'district']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
    ])

x_prapared = full_pipeline.fit_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression

LR_model = LinearRegression()

LR_model.fit(x_prapared, y)
LinearRegression()
```

```python
test_data = x_train.sample(10)
test_data
```

```python
y_labels = y.loc[test_data.index]
```

```python
test_data_prepared = full_pipeline.transform(test_data)
```

```python
predicted_data = LR_model.predict(test_data_prepared)
```

```python
pd.DataFrame({'Prognoz' : predicted_data, 'real baxosi' : y_labels})
```
